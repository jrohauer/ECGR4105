```python
# Josh Rohauer
# 8007098398
# Homework 1


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

#Regression
#Problem 1: Gradient descent for each
#Y = theta_0 + theta_1*X1
#Y = theta_0 + theta_2*X2
#Y = theta_0 + theta_3*X3
```

```python
dataset = pd.read_csv("/content/sample_data/D3.csv",header=0)
```

```python
#parameters
alpha = .01
theta_0 = 0
theta_1 = 0
theta_2 = 0
theta_3 = 0
iterations = 2000
```

```python
#calcs
n = len(dataset)-1
iter = np.arange(0,iterations,1)
reg_range=np.linspace(0,4,100)
```

```python
#Cost Function iteration
def J(X,Y):
  a = alpha
  t_0=theta_0
  t_1=theta_1

  for i in iter:
    Y_hat = t_0 + t_1*X
    cost = ((dataset.Y-Y_hat)**2).sum()/n
    loss.append(cost)

    dJdtheta_1 = ((X*(Y-Y_hat)).sum())*(-2/n)
    t_1 = t_1 - a*dJdtheta_1

    dJdtheta_0 = ((Y-Y_hat).sum())*(-2/n)
    t_0 = t_0 - a*dJdtheta_0

  return t_0, t_1
```

```python
loss=[]
t0, t1 = J(dataset.X1,dataset.Y)
Y_hat = t0 + t1*dataset.X1
cost = ((dataset.Y-Y_hat)**2).sum()/n
print("Y_hat = "+str(round(t0,3))+" + "+str(round(t1,3))+"*X1")
print("Cost = "+str(round(cost,5)))

Y_1 = t0+t1*reg_range
fig1, ax = plt.subplots()
ax.scatter(dataset.X1,dataset.Y)
ax.plot(reg_range,Y_1,'r')
ax.legend(['X1','Regression'])
ax.set_xlabel('X1')
ax.set_ylabel('Y')

fig1, ax1 = plt.subplots()
ax1.plot(loss)
ax1.legend(['Loss'])
ax1.set_xlabel('Iteration')
ax1.set_ylabel('Loss')
```

```python
loss=[]
t0, t1 = J(dataset.X2,dataset.Y)
Y_hat = t0 + t1*dataset.X2
cost = ((dataset.Y-Y_hat)**2).sum()/n
print("Y_hat = "+str(round(t0,3))+" + "+str(round(t1,3))+"*X2")
print("Cost = "+str(round(cost,5)))

Y_2 = t0+t1*reg_range
fig2, ax2 = plt.subplots()
ax2.scatter(dataset.X2,dataset.Y)
ax2.plot(reg_range,Y_2,'r')
ax2.legend(['X2','Regression'])
ax2.set_xlabel('X2')
ax2.set_ylabel('Y')

fig2, ax3 = plt.subplots()
ax3.plot(loss)
ax3.legend(['Loss'])
ax3.set_xlabel('Iteration')
ax3.set_ylabel('Loss')
```

```python
loss=[]
t0, t1 = J(dataset.X3,dataset.Y)
Y_hat = t0 + t1*dataset.X3
cost = ((dataset.Y-Y_hat)**2).sum()/n
print("Y_hat = "+str(round(t0,3))+" + "+str(round(t1,3))+"*X3")
print("Cost = "+str(round(cost,5)))

Y_3 = t0+t1*reg_range
fig3, ax4 = plt.subplots()
ax4.scatter(dataset.X3,dataset.Y)
ax4.plot(reg_range,Y_3,'r')
ax4.legend(['X3','Regression'])
ax4.set_xlabel('X3')
ax4.set_ylabel('Y')

fig3, ax5 = plt.subplots()
ax5.plot(loss)
ax5.legend(['Loss'])
ax5.set_xlabel('Iteration')
ax5.set_ylabel('Loss')
```

```python
#Problem 2
#Cost Function 3 variable iteration
def J3(X1,X2,X3,Y):
  a = alpha
  t_0=theta_0
  t_1=theta_1
  t_2=theta_2
  t_3=theta_3

  for i in iter:
    Y_hat = t_0 + t_1*X1 + t_2*X2 + t_3*X3
    cost = ((dataset.Y-Y_hat)**2).sum()/n
    loss.append(cost)

    dJdtheta_1 = ((X1*(Y-Y_hat)).sum())*(-2/n)
    t_1 = t_1 - a*dJdtheta_1

    dJdtheta_2 = ((X2*(Y-Y_hat)).sum())*(-2/n)
    t_2 = t_2 - a*dJdtheta_2

    dJdtheta_3 = ((X3*(Y-Y_hat)).sum())*(-2/n)
    t_3 = t_3 - a*dJdtheta_3

    dJdtheta_0 = ((Y-Y_hat).sum())*(-2/n)
    t_0 = t_0 - a*dJdtheta_0

  return t_0, t_1, t_2, t_3
```

```python
loss=[]
t0, t1, t2, t3 = J3(dataset.X1,dataset.X2,dataset.X3,dataset.Y)
Y_hat = t0 + t1*dataset.X1+ t2*dataset.X2+ t3*dataset.X3
cost = ((dataset.Y-Y_hat)**2).sum()/n
print("Y_hat = "+str(round(t0,3))+" + "+str(round(t1,3))+"*X1"+" + "+str(round(t2,3))+"*X2"+" + "+str(round(t3,3))+"*X3")
print("Cost = "+str(round(cost,5)))


fig4, ax6 = plt.subplots()
ax6.plot(loss)
ax6.legend(['Loss'])
ax6.set_xlabel('Iteration')
ax6.set_ylabel('Loss')
```
