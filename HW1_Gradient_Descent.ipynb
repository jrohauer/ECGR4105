```python
file_path = '/content/drive/My Drive/ML_data/D3.csv'
dataset = pd.read_csv(file_path)
```

```python
#parameters
alpha = .01
theta2 = np.zeros(2).reshape(2,1)
theta4 = np.zeros(4).reshape(4,1)
iterations = 2000
```

```python
#calcs
n = len(dataset)-1
iter = np.arange(0,iterations,1)
reg_range=np.linspace(0,4,100)
Y = dataset.values[:, 3].reshape(n+1,1)
X1 = dataset.values[:, 0].reshape(n+1,1)
X2 = dataset.values[:, 1].reshape(n+1,1)
X3 = dataset.values[:, 2].reshape(n+1,1)
X0 = np.ones((n+1,1))
loss=[]
```

```python
#Cost Function iteration
def J(X,Y):
  a = alpha
  theta = theta2

  for i in iter:
    Y_hat = X.dot(theta).reshape(n+1,1)
    errors = np.subtract(Y_hat,Y)
    sqrerrors = np.square(errors)
    cost = 1/(2*n+1) * sqrerrors.sum()
    loss.append(cost)


    sum_delta = (a / (n+1)) * X.transpose().dot(errors)
    theta -= sum_delta

  return theta
```

```python
X = np.hstack((X0,X1))
theta = J(X,Y)
Y_hat = X.dot(theta)
cost = loss[-1]
print("Y_hat = "+str(round(float(theta[0]),3))+" + "+str(round(float(theta[1]),3))+"*X1")
print("Cost = "+str(round(cost,5)))

Y_1 = float(theta[0])+float(theta[1])*reg_range
fig1, ax = plt.subplots()
ax.scatter(X1,Y)
ax.plot(reg_range,Y_1,'r')
ax.legend(['X1','Regression'])
ax.set_xlabel('X1')
ax.set_ylabel('Y')

fig1, ax1 = plt.subplots()
ax1.plot(loss)
ax1.legend(['Loss'])
ax1.set_xlabel('Iteration')
ax1.set_ylabel('Loss')
```
